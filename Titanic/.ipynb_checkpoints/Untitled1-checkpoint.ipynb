```python
def custom_split(str):
    if str[0].isnumeric():
        return str
    else:
        a=0
        i=len(str)
        while(i>1):
            if str[i-1]== " ":
                a = len(str)-i+1
                break
            i = i-1
        str1 = str[:-a:-1]
        return str1[::-1]
    
custom_split("541")
```
