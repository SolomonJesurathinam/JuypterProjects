## Variables

```python
#Boolean variable
boolean = True
boolean1 = bool(0)
print(type(boolean))
print(boolean1,type(boolean1))
```

```python
#list variable
list_variable = [1,2,3,12.3]
type(list_variable)
```

```python
list_variable[0] = 123
list_variable
```

```python
#Tuple is immutable - you cant change value
tuple_variable = (1,2,3,4,4)
type(tuple_variable)
```

```python
tuple_variable[0] = 123
tuple_variable
```

```python
#set variable
set_variable = {1,2,3,4,5}
type(set_variable)
```
